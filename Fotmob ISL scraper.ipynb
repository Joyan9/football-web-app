```python
def get_driver():
    chrome_options = Options()
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--disable-dev-shm-usage')
    # loading the browser
#     driver = webdriver.Chrome("chromedriver",options=chrome_options)
    driver = webdriver.Chrome(ChromeDriverManager().install(), options=chrome_options)
    return driver
```

```python
df1 = pd.read_csv("ISL_scraped_stats_part1.csv")
df2 = pd.read_csv("ISL_scraped_stats_part2.csv")
df = pd.concat([df1,df2],ignore_index=True)
```

```python
df['xG_differential'] = df['home_xG'] - df['away_xG']
df.head(3)
```

```python
for idx, row in df.iterrows():
    if df.loc[idx,'xG_differential']>1.5:
        df.loc[idx,'home_xP'],df.loc[idx,'away_xP'] = 2.7,0.3
    elif df.loc[idx,'xG_differential']>1.0 and df.loc[idx,'xG_differential']<1.5:
        df.loc[idx,'home_xP'],df.loc[idx,'away_xP'] = 2.3,0.7
    
    elif df.loc[idx,'xG_differential']>0.5 and df.loc[idx,'xG_differential']<1.0:
        df.loc[idx,'home_xP'],df.loc[idx,'away_xP'] = 2.0,1.0
    
    elif df.loc[idx,'xG_differential']>0 and df.loc[idx,'xG_differential']<0.5:
        df.loc[idx,'home_xP'],df.loc[idx,'away_xP'] = 1.5,1.5
    
    elif df.loc[idx,'xG_differential']>-0.5 and df.loc[idx,'xG_differential']<0:
        df.loc[idx,'home_xP'],df.loc[idx,'away_xP'] = 0.7,2.3
    
    elif df.loc[idx,'xG_differential']>-1.0 and df.loc[idx,'xG_differential']<-0.5:
        df.loc[idx,'home_xP'],df.loc[idx,'away_xP'] = 0.5,2.5
        
    elif df.loc[idx,'xG_differential']>-1.5 and df.loc[idx,'xG_differential']<-1.0:
        df.loc[idx,'home_xP'],df.loc[idx,'away_xP'] = 0.3,2.7
    
    elif df.loc[idx,'xG_differential']>-1.5:
        df.loc[idx,'home_xP'],df.loc[idx,'away_xP'] = 0.1,2.9
        
    
```

```python
# team_xG_first_half,team_xG_second_half,team_xG_Open_Play,team_xG_Set_Play,team_xGOT
team_df = pd.merge(team_goals,team_xG,how='inner',on='team')
team_df = pd.merge(team_df,team_goals_conceded,how='inner',on='team')
team_df = pd.merge(team_df,team_xGA,how='inner',on='team')
team_df = pd.merge(team_df,team_shots,how='inner',on='team')
team_df = pd.merge(team_df,team_shots_against,how='inner',on='team')
team_df = pd.merge(team_df,team_xPoints,how='inner',on='team')
team_df = pd.merge(team_df,team_xG_first_half,how='inner',on='team')
team_df = pd.merge(team_df,team_xGA_first_half,how='inner',on='team')
team_df = pd.merge(team_df,team_xG_second_half,how='inner',on='team')
team_df = pd.merge(team_df,team_xGA_second_half,how='inner',on='team')
team_df = pd.merge(team_df,team_xG_Open_Play,how='inner',on='team')
team_df = pd.merge(team_df,team_xGA_Open_Play,how='inner',on='team')
team_df = pd.merge(team_df,team_xG_Set_Play,how='inner',on='team')
team_df = pd.merge(team_df,team_xGA_Set_Play,how='inner',on='team')
team_df = pd.merge(team_df,team_xGOT,how='inner',on='team')
team_df = pd.merge(team_df,team_xGOTA,how='inner',on='team')
team_df
```
